```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.formula1.com/en/results/2025/drivers'
page = requests.get(url)
print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
tables = soup.find_all('table')
# print(len(table))
for i, table in enumerate(tables):
    print(f"\n--- Table {i} ---")
    print(table.get_text()[:300])
```

```python
table = soup.find_all('table')[0]
titles = table.find_all('th')
# print(titles)
titles_row = [title.text.strip() for title in titles]
print (titles_row)
```

```python
rows = table.find_all('tr')
for row in rows:
    data_row = row.find_all('td')
    individual_data_row = [data.text.strip() for data in data_row]
    print(individual_data_row)
```

```python
data = []

for row_id, row in enumerate(rows):
    cols = row.find_all('td')
    # print(f"row = {row_id}: {cols}")
    if cols:
        Position = cols[0].text.strip()
        Driver = cols[1].text.strip()[:-3]
        Nationality = cols[2].text.strip()
        Team = cols[3].text.strip()
        Points = cols[4].text.strip()
        # print(Driver)
        data.append([Position, Driver, Nationality, Team, Points])

data
```

```python
df = pd.DataFrame(data, columns=titles_row)
```

```python
df.to_csv(r'formula1.csv', index=False, encoding='utf-8-sig')
```
