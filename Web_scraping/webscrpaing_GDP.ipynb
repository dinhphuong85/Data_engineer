```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

headers = {
    "User-Agent" : "Mozilla/5.0; (Windows NT 10.0; win64; x64)"
}

page = requests.get(url, headers = headers)
print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
tables = soup.find_all('table')
print(len(tables))
```

```python
for i, table in enumerate(tables):
    print(f"\n--- Table {i} ---")
    print(table.get_text()[:300])
```

```python
table = soup.find_all('table')[2]
print(table.get_text())
```

```python
titles = table.find_all('th')
titles_row = [title.text.strip() for title in titles]
print(titles_row)
```

```python
df = pd.DataFrame(columns = titles_row)
```

```python
rows = table.find_all('tr')
for row in rows[1:]:
    data_row = row.find_all('td')
    individual_data_row = [data.text.strip() for data in data_row]
    # print(individual_data_row)
    length = len(df)
    df.loc[length] = individual_data_row
```

```python
df.to_csv(r'list_countries_by_GDP.csv', index = False)
```
